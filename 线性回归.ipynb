```python
import regression
from numpy import *     #导入numpy中的全部函数
xArr,yArr = regression.loadDataSet('C:\\Users\\libo\\Desktop\\learn\\machinelearninginaction\\Ch08\\ex0.txt')    #相对·绝对位置
xArr[0:2]
```

```python
ws = regression.standRegres(xArr,yArr)        #计算w的值（矩阵形式的多值）
ws
```

```python
xMat = mat(xArr)
yMat = mat(yArr)
yHat = xMat * ws       #得出预测值Y帽
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()     #创建自定义图像
ax = fig.add_subplot(111)     #创建一个代号为111的子图
ax.scatter(xMat[:,1].flatten().A[0], yMat.T[:,0].flatten().A[0])             #创建图像并绘出原始数据
plt.show()
```

```python

fig = plt.figure() 
ax = fig.add_subplot(111)
xCopy = xMat.copy()      #复制xMat
xCopy.sort(0)      #按照升序排列
yHat = xCopy * ws      #预测的Y值
ax.plot(xCopy[:,1] , yHat)
plt.show()
```
